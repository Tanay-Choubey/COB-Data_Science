### Problem Statement: Create a csv dataset using python , pandas and any public api.

### Dataset : https://api.themoviedb.org/3/movie/top_rated?api_key=aaa7de53dcab3a19afed86880f364e54&language=en-US&page=1

### Importing the neccesary Libraries

```python
import pandas as pd
import requests
```

### Requesting to the api

```python
response = requests.get('https://api.themoviedb.org/3/movie/top_rated?api_key=aaa7de53dcab3a19afed86880f364e54&language=en-US&page=1')
```

```python
# Creating a DataFrame
df = pd.DataFrame()
```

### Checking if the request is successful

```python
if response.status_code == 200:
	for i in range(1, 400):
		response = requests.get('https://api.themoviedb.org/3/movie/top_rated?api_key=aaa7de53dcab3a19afed86880f364e54&language=en-US&page=1'.format(i))
		temp_df = pd.DataFrame(response.json()['results'])[['id','title', 'overview', 'release_date', 'popularity','vote_average', 'vote_count']]
		df = df.append(temp_df, ignore_index=True)
else:
	print('Error', response.status_code)
```

### Saving data to csv

```python
df.to_csv('movie_rating.csv', index=False)
```
